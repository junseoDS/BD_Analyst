# 3. 이상치 처리

```python
import pandas as pd
data= pd.read_csv('EX_CEOSalary.csv')
```

```python
data.head()
```

```python
data.boxplot(column='salary')
```

```python
data.boxplot(column='sales')
```

```python
data.boxplot(column='roe')
```

```python
data['salary'].mean()
Q1_salary = data['salary'].quantile(q=0.25)
Q3_salary = data['salary'].quantile(q=0.75)

```

```python
IQR_salary = Q3_salary-Q1_salary
IQR_salary
```

```python
data_IQR = data[(data['salary']<Q3_salary+1.5*IQR_salary)&(data['salary']>Q1_salary-1.5*IQR_salary)]
```

```python
data_IQR['salary'].hist()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_IQR['roe'],data_IQR['salary'],alpha=0.5)
```

## 2.Sales와  Salary 모두 이상치 처리한 데이터셋

```python
Q1_sales = data['sales'].quantile(q=0.25)
Q3_sales = data['sales'].quantile(q=0.75)
IQR_sales = Q3_sales - Q1_sales
IQR_sales
```

```python
data_IQR = data[(data['salary']<Q3_salary + IQR_salary*1.5)&(data['salary']>Q1_salary - IQR_salary*1.5)&
               (data['sales']<Q3_sales+IQR_sales*1.5)&(data['sales']>Q1_sales-IQR_sales*1.5)]
```

```python
data_IQR.hist(bins=50)
```

```python
data_IQR.corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_IQR['roe'],data_IQR['salary'],alpha=0.5)
```

# 6. 데이터 정제 실전


```python
import pandas as pd
data = pd.read_csv('house_raw.csv')
data
```

```python
data.describe()
```

```python
data.hist(bins=50)
```

```python
X = data[data.columns[0:5]]
y = data[['house_value']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_minmax = MinMaxScaler()
```

```python
scaler_minmax.fit(X_train)
```

```python
scaler_minmax_train = scaler_minmax.transform(X_train)
scaler_minmax_test = scaler_minmax.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(scaler_minmax_train,y_train)
```

```python
pred_train =model.predict(scaler_minmax_train)
```

```python
model.score(scaler_minmax_train,y_train)
```

```python
pred_test =model.predict(scaler_minmax_test)
model.score(scaler_minmax_test,y_test)
```
