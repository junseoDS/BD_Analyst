```python
import pandas as pd
data = pd.read_csv('Fvote.csv')
data.head()
```

```python
X = data[data.columns[1:13]]
y=data[['vote']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
```

# 1. 그리드 서치

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.001,0.01,0.1,1,10,100]}
```

```python
grid_search = GridSearchCV(LogisticRegression(),param_grid,cv=5,return_train_score=True)
```

```python
grid_search.fit(X_train,y_train)
```

```python
print("최적의 파라미터는 ", grid_search.best_params_)
print("최적의 정확도는 ", grid_search.best_score_)
```

```python
result_grid=pd.DataFrame(grid_search.cv_results_)
```

```python
import matplotlib.pyplot as plt
plt.plot(result_grid['param_C'],result_grid['mean_train_score'],label='Train')
plt.plot(result_grid['param_C'],result_grid['mean_test_score'],label='Test')
plt.legend()
```

# 2. 랜덤 서치

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_distribs = {'C':randint(1,100)}
```

```python
random_search = RandomizedSearchCV(LogisticRegression(),param_distributions=param_distribs,cv=5,
                                  return_train_score=True)
```

```python
random_search.fit(X_train,y_train)
```

```python
print("최적의 파라미터는 ", random_search.best_params_)
print("최적의 정확도는 ", random_search.best_score_)
```

```python
random_search.score(X_test,y_test)
```

```python
result_random=pd.DataFrame(random_search.cv_results_)
```

```python
result_random
```

```python
import matplotlib.pyplot as plt
plt.plot(result_random['param_C'],result_random['mean_train_score'],label='Train')
plt.plot(result_random['param_C'],result_random['mean_test_score'],label='Test')
plt.legend()
```
